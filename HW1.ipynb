```python
import sys
import sklearn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge,Lasso
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.linear_model import RidgeCV, LassoCV
import warnings

%matplotlib inline
warnings.filterwarnings('ignore')
```

## Data pre-processing

First,see what the data look like

```python
df =  pd.read_csv('Hitters.csv',index_col=0)
df.head()
```

Then see the statistic feathers of the data, and we find 'Salary' (which is our label) column has a lot NANs.

```python
print(df.describe())
```

Then I dropped all these missing-label data and choose all other numeric feathers as X and 'Salary' as Y

```python
df = df.dropna()
X = df.drop(columns=['League','Division','NewLeague','Salary'])
Y = df.Salary
```

Split the data for training and testing. 70% for training and 30% for testing.

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3)
```

## 1 Regularization

### 1.1 LASSO MODEL

First, we fit the Lasso model (whose alpha is by default 1.0)and predict on X_test. The graph is visulization of predict result and ground-true value (Y_test).

Red points stand for predict

Blue points stand for ground-true

```python
lasso = Lasso(normalize=True)
lasso.fit(X_train,Y_train)
lasso_result = lasso.predict(X_test)

plt.plot(lasso_result,alpha=0.7,marker='*',markersize=5,color='red',label=r'predict')
plt.plot(Y_test,alpha=0.7,marker='d',markersize=5,color='blue',label=r'groundtrue')

plt.xticks([])

plt.xlabel('X',fontsize = 16)
plt.ylabel('Y',fontsize = 16)
```

Then create a visualization of the coefficient trajectories. And the final three predictors are [Hits, CHits, CRBI] (Check which weights are not zero, but I since everytime the regressor will change, so the final three predictors are not always the same.)

```python
alphas = [0.0001,0.001,0.01,0.1,1,10,100,1000,10000]
coefs = []
for a in alphas:
    lasso = Lasso(alpha=a,normalize=True)
    lasso.fit(X_train,Y_train)
    coefs.append(lasso.coef_)
coefs_arr = np.array(coefs).T

#Reference 'to https://www.cnblogs.com/pinard/p/6023000.html'
pic = plt.gca()
pic.plot(alphas, coefs)
pic.set_xscale('log') 
plt.xlabel('alpha')
plt.ylabel('weights')
plt.title('coefficients trajectories')
plt.show()
```

Then I use cross-validation get the best alpha and remaining preditor number can be seen.

```python
lassocv = LassoCV(normalize = True)
lassocv.fit(X_train,Y_train)
print('Best alpha is:',lassocv.alpha_)
count = 0
for i in lassocv.coef_:
    if i != 0:
        count = count + 1
print('There are %d preditors left'%count)
```

### 1.2 RIDGE MODEL

First, we visualized the coefficients trajectories.

```python
alphas = [0.0001,0.001,0.01,0.1,1,10,100,1000,10000]
coefs = []
for a in alphas:
    rr = Ridge(alpha=a,normalize=True)
    rr.fit(X_train,Y_train)
    coefs.append(rr.coef_)
coefs_arr = np.array(coefs).T

#Reference 'to https://www.cnblogs.com/pinard/p/6023000.html'
pic = plt.gca()
pic.plot(alphas, coefs)
pic.set_xscale('log') 
plt.xlabel('alpha')
plt.ylabel('weights')
plt.title('coefficients trajectories')
plt.show()
```

Then we find the optimal value of alpha is 0.01 for this case.

### 2.1 Explain in your own words the bias-variance tradeoff

Bias is the difference between the final result we get and the real result. Variance is the instability of the results we get ourselves. These two concepts are for the model. When our model is relatively simple, the bias obtained will be relatively high. But because it is simple, it is stable, so the variance will be lower. When our model is gradually complicated, the bias will gradually decline, but there will be a loss in stability, so the variance will increase. Negative correlation between the two is like time and space, so we have to make a trade-off.


### 2.2 What role does regularization play in this tradeoff?

In practice, we often encounter over-fitting situations. We want our model reach a balance point of bias and variance, so when facing over-fitting problem, we use regularization to properly increase bias then we decrease the variance and approach the balance point.

### 2.3 Make reference to your findings in number (1) to describe models of high/low bias and variance.

The difference between Lasso and Ridge is the way they add the penalty. Lasso uses L1 regularization while Ridge uses L2 regularization. What is the same between them is that compare to linear regression, Lasso and Ridge regression both add a penalty to restrict coefficients. By increasing penalty, we can shrink the value of coefficients and then decrease the variance to prevent over-fitting problem. Though in the same time bias will increase, we still can use this to reach a respectively balanced point.
